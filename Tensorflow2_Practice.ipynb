```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
x = df.drop("Survived", axis=1)
x
```

```python
y = df["Survived"]
y
```

```python
print("y is {} datatype.".format(np.dtype(y)))
print("x has {} shape.".format(x.shape))
print("y has {} shape.".format(y.shape))
```

```python
#to see how many null value in data set
x.isnull().sum()
```

```python
#dropping Cabin as it has many null values in it
x = x.drop("Cabin", axis=1)
```

```python
x.describe()
```

```python
#x.describe(include=['O'])
```

```python
# filling the missing values
x['Age'].fillna(x.groupby("Pclass")["Age"].transform("median"), inplace=True)
x["Embarked"].fillna("S", inplace = True)
#x.loc[x.Embarked.isnull(), 'Embarked'] = 'S'
```

```python
x.isnull().sum() #Now, we dont have any null value
```

```python
x = x.drop("Ticket", axis=1)
x = x.drop("Name", axis=1)
x = x.drop("PassengerId", axis=1)
# dropping these column as it is not so much relevant for us
```

```python
x.head()
```

```python
# mapping the string to integer 
x["Sex"] = x['Sex'].map( {'female': 0, 'male': 1} ).astype(int)
x["Embarked"] = x['Embarked'].map({'S': 0, 'C': 1, 'Q': 2}).astype(int)
```

```python
x.head()
```

```python
import tensorflow as tf
#converting every column typecast to float32
x["Age"] = tf.cast(x["Age"], tf.float32)
x["Pclass"] = tf.cast(x["Pclass"], tf.float32)
x["Sex"] = tf.cast(x["Sex"], tf.float32)
x["Parch"] = tf.cast(x["Parch"], tf.float32)
x["Fare"] = tf.cast(x["Fare"], tf.float32)
x["Embarked"] = tf.cast(x["Embarked"], tf.float32)
x["SibSp"] = tf.cast(x["SibSp"], tf.float32)
```

```python
x.head()
```

```python
# converting dataset to array type
x = np.array(x)
y = np.array(y)
```

```python
from sklearn.model_selection import train_test_split 
# create training and testing vars
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2)
print(train_x.shape, train_y.shape)
print(test_x.shape, test_y.shape)
```

```python
titanicmodel = tf.keras.models.Sequential([
tf.keras.layers.Flatten(),
tf.keras.layers.Dense(512,activation=tf.nn.relu),
tf.keras.layers.Dropout(0.2),
tf.keras.layers.Dense(140,activation=tf.nn.elu),
tf.keras.layers.Dense(2,activation=tf.nn.softmax)
])
```

```python
optimiser = tf.keras.optimizers.Adam()
titanicmodel.compile (optimizer= optimiser, loss='sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
titanicmodel.fit(train_x, train_y, batch_size=60, epochs=25)
```

```python
titanicmodel.evaluate(test_x, test_y)
```
